```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
```

```python
file = '/kaggle/input/zomato-reviews-ratings/zomato_reviews.csv'
df = pd.read_csv(file,usecols=['rating','review'])
print(f"Found {df.shape[0]} reviews with {df.rating.nunique()} ratings..")
```

```python
df.head()
```

```python
def plot_cloud(rating):
    temp = df[df['rating'] == rating]
    unique_string=(" ").join(temp['review'])
    wordcloud = WordCloud(width = 1500
                          ,height = 1000
                          ,background_color='white'
                          ,min_word_length=4
                         ,stopwords=stops
                          ,colormap ='hot').generate(unique_string)
    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

# Prominent words in rating=5 reviews

```python
plot_cloud(5)
```

# Prominent words in rating=1 reviews

```python
plot_cloud(1)
```
